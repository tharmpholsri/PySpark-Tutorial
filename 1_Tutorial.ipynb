### DDL SCHEMA

```python
my_ddl_schema = '''
  Item_Identifier string,
  Item_Weight string,
  Item_Fat_Content string,
  Item_Visibility double,
  Item_Type string,
  Item_MRP double,
  Outlet_Identifier string,
  Outlet_Establishment_Year integer,
  Outlet_Size string,
  Outlet_Location_Type string,
  Outlet_Type string,
  Item_Outlet_Sales double'''
```

```python
df = spark.read.format('csv')\
    .schema(my_ddl_schema)\
        .option('header',True)\
            .load('/Volumes/workspace/default/storage/BigMart Sales.csv')
```

```python
df.display()
```

```python
df.printSchema()
```

### StructType() Schema

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```python
my_struct_schema = StructType([
    StructField('Item_Identifier', StringType(),True),
    StructField('Item_Weight', StringType(),True),
    StructField('Item_Fat_Content', StringType(),True),
    StructField('Item_Visibility', StringType(),True),
    StructField('Item_Type', StringType(),True),
    StructField('Item_MRP', StringType(),True),
    StructField('Outlet_Identifier', StringType(),True),
    StructField('Outlet_Establishment_Year', StringType(),True),
    StructField('Outlet_Size', StringType(),True),
    StructField('Outlet_Location_Type', StringType(),True),
    StructField('Outlet_Type', StringType(),True),
    StructField('Item_Outlet_Sales', StringType(),True)
])
```

```python
df = spark.read.format('csv')\
    .schema(my_struct_schema)\
        .option('header',True)\
            .load('/Volumes/workspace/default/storage/BigMart Sales.csv')
```

```python
df.display()
```

```python
df.printSchema()
```

### SELECT

```python
df = spark.read.format('csv').option('inferSchema','true').option('header','true').load('/Volumes/workspace/default/storage/BigMart Sales.csv')
```

```python
df.display()
```

```python
df_sel = df.select('Item_Identifier','Item_Weight','Item_Fat_Content').display()
```

```python
df_sel = df.select(col('Item_Identifier'),col('Item_Weight'),col('Item_Fat_Content')).display()
```

### ALIA

rename a column

```python
df.select(col("Item_Identifier").alias('Item_ID')).display()
```

### FITER

```python
df.display()
```

#### Scenario - 1

show only rows that have Regular in column Item_Fat_column

```python
df.filter(col('Item_Fat_Content') == 'Regular').display()
```

#### Scenorio - 2

```python
df.filter((col('Item_Type') == 'Soft Drinks') & (col('Item_Weight') < 10)).display()
```

#### Scenorio - 3

```python
df.filter((col('Outlet_Size').isNull()) & (col('Outlet_Location_Type').isin('Tier 1','Tier 2'))).display()
```

#### withColumnRenamed

```python
df.withColumnRenamed('Item_Weight','Item_Wt').display()
```

### withColumn

create new column and modify column

#### Scenario - 1

```python
df = df.withColumn('flag', lit("new")).display()
```

```python
df.withColumn('multiply', col('Item_Weight')*col('Item_MRP')).display()
```

#### Scenario - 2

```python
df.withColumn('Item_Fat_Content',regexp_replace(col('Item_Fat_Content'),"Regular", "Reg"))\
    .withColumn('Item_Fat_Content',regexp_replace(col('Item_Fat_Content'),"Low Fat", "Lf")).display()
```

#### Type Casting

```python
df = df.withColumn('Item_Weight', col('Item_Weight').cast(StringType()))
```

```python
df.printSchema()
```
